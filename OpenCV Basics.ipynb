```python
#Object Detection (Haarcascascades Classifier)
facesImg = cv2.imread('kidsimg.jpg')
facesImg = cv2.cvtColor(facesImg, cv2.COLOR_BGR2RGB)
plt.imshow(facesImg)
plt.axis('off')
plt.show()
```

```python
model = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces = model.detectMultiScale(facesImg,1.3,1)
```

```python
faces.shape
```

```python
for f in faces:
    x,y,w,h = f
    cv2.rectangle(facesImg,(x,y),(x+w,y+h),(0,255,0),2)

plt.imshow(facesImg)
plt.show()
```
